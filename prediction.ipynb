```python
from huggingface_hub import from_pretrained_keras
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Load the BiLSTM model
model_name = "aya2003/my_bilstm_model"
loaded_model = from_pretrained_keras(model_name)

# Sample input text for inference
text = "هذا نص عربي للاختبار"

# Example word_index (Replace with your actual word_index from training)
word_index = {"هذا": 1, "نص": 2, "عربي": 3, "للاختبار": 4}  # This is just an example

# Convert the text to a sequence of word indices
sequence = [word_index.get(word, 0) for word in text.split()]

# Pad the sequence to match the input length expected by the BiLSTM model
MAX_SEQUENCE_LENGTH = 100  # Replace with the max length used during training
padded_sequence = pad_sequences([sequence], maxlen=MAX_SEQUENCE_LENGTH, padding='post')

# Convert to numpy array
padded_sequence = np.array(padded_sequence)

# Make a prediction
predictions = loaded_model.predict(padded_sequence)

# Get the class with the highest probability
predicted_class = np.argmax(predictions, axis=-1)

print(f"Predicted label: {predicted_class[0]}")
```
